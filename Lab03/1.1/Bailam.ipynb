## 1.1.1. Ôn tập lý thuyết

---

### Thống kê mô tả là gì?
- **Khái niệm:** Thống kê mô tả (Descriptive Statistics) là tập hợp các phương pháp dùng để **tóm tắt, trình bày và mô tả dữ liệu** thu thập được.  
- **Khác với thống kê suy luận (Inferential Statistics):**
  | Tiêu chí | Thống kê mô tả | Thống kê suy luận |
  |-----------|----------------|-------------------|
  | Mục đích | Mô tả dữ liệu hiện có | Suy luận cho toàn bộ quần thể |
  | Dữ liệu | Mẫu hoặc toàn bộ tập dữ liệu | Mẫu đại diện |
  | Kết quả | Bảng, biểu đồ, giá trị trung bình... | Kiểm định giả thuyết, ước lượng, dự đoán |

---

###  Các thước đo thống kê mô tả chính
- **Trung bình (Mean):** giá trị đại diện cho dữ liệu.
- **Trung vị (Median):** giá trị giữa của dữ liệu đã sắp xếp.
- **Phương sai (Variance)** và **Độ lệch chuẩn (Standard Deviation):** đo mức độ phân tán.
- **Khi nào nên dùng trung vị thay vì trung bình?**
  → Khi dữ liệu có **ngoại lệ (outlier)** hoặc **phân bố lệch**.

---

###  Xác định phân bố của tập dữ liệu
- Các loại phân bố phổ biến:
  - **Phân bố chuẩn (Normal)**
  - **Phân bố lệch trái (Left-skewed)**
  - **Phân bố lệch phải (Right-skewed)**
- **Cách xác định:**  
  - Quan sát **histogram** hoặc **boxplot**  
  - Tính **hệ số Skewness, Kurtosis**

---

###  Độ lệch chuẩn và phạm vi (Range)
- **Phạm vi (Range)** = Giá trị lớn nhất − nhỏ nhất  
- **Độ lệch chuẩn (SD)** cho biết dữ liệu **phân tán quanh trung bình** đến mức nào  
  → Càng nhỏ ⇒ dữ liệu càng ổn định

---

###  Phân vị Q1, Q2, Q3 trong Boxplot
| Thước đo | Ý nghĩa | Vị trí |
|-----------|----------|--------|
| Q1 | Phân vị thứ 25% | đáy hộp |
| Q2 | Trung vị (50%) | đường giữa hộp |
| Q3 | Phân vị thứ 75% | đỉnh hộp |

---

###  Xử lý giá trị thiếu (Missing values)
- **Cách xử lý:**
  - Xóa dòng có giá trị thiếu (`dropna()`)
  - Thay thế bằng trung bình/trung vị (`fillna(mean)` hoặc `fillna(median)`)
  - Dự đoán giá trị thiếu (imputation)

---

###  Đọc và diễn giải biểu đồ
- **Histogram:** cho biết **phân bố tần suất** của dữ liệu  
- **Boxplot:** thể hiện **độ phân tán**, **ngoại lệ (outlier)** và **phân vị**

---

###  Xử lý giá trị ngoại lai (Outliers)
- **Phát hiện:** bằng Boxplot, Z-score hoặc IQR  
- **Xử lý:**  
  - Giữ nguyên (nếu có ý nghĩa thực tế)  
  - Loại bỏ  
  - Winsorize (giới hạn giá trị ở ngưỡng nhất định)

- 1. Tính mean, median, mode, variance, standard deviation, range, percentile, quartile, interquartile range (IQR) sử dụng thư viện numpy và stats trên tập dữ liệu COVID.

```python
import numpy as np 
import pandas as pd
from scipy import stats
```

```python
# Load the .csv into a dataframe using read_csv 
covid_data = pd.read_csv("covid-data.csv") 
covid_data = covid_data[['iso_code','continent', 
                 'location','date','total_cases','new_cases']]
```

```python
# Take a quick look at the data 
covid_data.head(5) 
covid_data.dtypes 
covid_data.shape
```

```python
# Get the mean of the data 
data_mean = np.mean(covid_data["new_cases"]) 
# Get the median of the data 
data_median = np.median(covid_data["new_cases"]) 
# Get the mode of the data 
data_mode = stats.mode(covid_data["new_cases"]) 
# Obtain the variance of the data 
data_variance = np.var(covid_data["new_cases"]) 
# Obtain the standard deviation of the data 
data_sd = np.std(covid_data["new_cases"]) 
# Compute the maximum and minimum values of the data 
data_max = np.max(covid_data["new_cases"]) 
data_min = np.min(covid_data["new_cases"]) 
# Obtain the 60th percentile of the data 
data_percentile = np.percentile(covid_data["new_cases"],60) 
# Obtain the quartiles of the data 
data_quartile = np.quantile(covid_data["new_cases"],0.75) 
# Get the IQR of the data 
data_IQR = stats.iqr(covid_data["new_cases"])
```

# Wine Quality (Red Wine) – Exploratory Data Analysis (EDA)

## 1. Giới thiệu

Bộ dữ liệu **Wine Quality - Red Wine** cung cấp các thông tin hóa học của rượu vang đỏ,  
mục tiêu là **đánh giá chất lượng rượu (quality)** theo thang điểm 0–10 do chuyên gia chấm.

**Nguồn:** [UCI Machine Learning Repository - Wine Quality Dataset](https://www.kaggle.com/code/eisgandar/red-wine-quality-eda-classification)

**Mục tiêu notebook:**
- Làm sạch dữ liệu  
- Thống kê mô tả các biến  
- Phân tích mối quan hệ giữa các đặc trưng và chất lượng rượu


1. Import thư viện và các dữ liệu vào notebook

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc dữ liệu
wine_data = pd.read_csv("winequality-red.csv")

# Hiển thị 5 dòng đầu
wine_data.head()
```

2. Kiểm tra thông tin tổng quan
Xem thông tin tổng quan:
- Kiểu dữ liệu  
- Thống kê mô tả  
- Kiểm tra dữ liệu thiếu


```python
# Thông tin cấu trúc dữ liệu
wine_data.info()

# Kiểm tra giá trị trùng lặp
duplicate_rows = wine_data.duplicated().sum()
print(f"Số dòng trùng lặp: {duplicate_rows}")

# Loại bỏ dòng trùng lặp (nếu có)
wine_data = wine_data.drop_duplicates()
```

3. Kiểm tra và xử lý dữ liệu thiếu

```python
# Kiểm tra giá trị bị thiếu
print("Số lượng giá trị thiếu trên mỗi cột:")
print(wine_data.isnull().sum())

# Nếu có giá trị thiếu, có thể điền trung bình hoặc loại bỏ (ở đây: dropna)
wine_data = wine_data.dropna()
wine_data.shape
```

4. Thống kê mô tả

```python
# Thống kê các chỉ số mô tả
wine_data.describe().T
```

5. Phân phối và tần suất biến chất lượng

```python
sns.countplot(x='quality', data=wine_data, hue='quality', palette='coolwarm', legend=False)
plt.title('Phân phối điểm chất lượng rượu')
plt.xlabel('Điểm chất lượng')
plt.ylabel('Số lượng mẫu')
plt.show()

# Tần suất chi tiết
wine_data['quality'].value_counts().sort_index()
```

6. Phân phối các biến định lượng (Histogram)

```python
wine_data.hist(figsize=(12, 10), bins=15, color='skyblue', edgecolor='black')
plt.suptitle('Phân phối các biến đặc trưng trong dữ liệu Wine Quality')
plt.show()
```

7. Ma trận tương quan mô tả xu hướng giữa các biến

```python
plt.figure(figsize=(10, 8))
sns.heatmap(wine_data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Ma trận tương quan giữa các biến (Descriptive)')
plt.show()
```

# Pima Indians Diabetes – Exploratory Data Analysis (EDA)

---

## 1. Giới thiệu

Bộ dữ liệu **Pima Indians Diabetes Dataset** chứa thông tin y tế của phụ nữ gốc Pima (Mỹ).  
Mục tiêu: **Dự đoán xem người đó có bị tiểu đường hay không** dựa trên các chỉ số lâm sàng.

**Nguồn:** [Kaggle - Pima Indians Diabetes Database](https://www.kaggle.com/datasets/uciml/pima-indians-diabetes-database)

**Mục tiêu notebook:**
- Làm sạch và xử lý dữ liệu  
- Khám phá mối quan hệ giữa các biến  
- Chuẩn bị dữ liệu cho mô hình học máy


1. Import thư viện và nạp dữ liệu vào notebook

```python
# Import thư viện 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
sns.set_theme()

# Nạp dữ liệu
df = pd.read_csv("diabetes.csv")

# Xem kích thước và 5 dòng đầu
print("Kích thước dữ liệu:", df.shape)
df.head()
```

2. Tổng quan dữ liệu

  Hiển thị thông tin cơ bản để nắm được cấu trúc dataset:
- 5 dòng đầu tiên
- Thông tin kiểu dữ liệu
- Mô tả thống kê


- `Outcome = 0` → Người **không mắc tiểu đường**
- `Outcome = 1` → Người **mắc tiểu đường**

Kiểm tra sự cân bằng giữa hai nhóm bằng biểu đồ cột.


```python
#Thông tin tổng quan về dữ liệu
df.info()

#Biến mục tiêu
sns.countplot(x='Outcome', data=df, hue='Outcome', palette='coolwarm', legend=False)
plt.title('Phân phối biến mục tiêu (Outcome)')
plt.xlabel('Outcome (1 = Có tiểu đường, 0 = Không)')
plt.ylabel('Số lượng mẫu')
plt.show()

print("Tần suất Outcome:")
print(df['Outcome'].value_counts(normalize=True).mul(100).round(2))
```

```python
# Đếm số lượng mỗi giá trị trong cột 'Outcome'
target_counts = df['Outcome'].value_counts()
target_percent = df['Outcome'].value_counts(normalize=True) * 100

print("Target counts:\n", target_counts)
print("\nTarget percentage:\n", target_percent)
```

Một số cột chứa giá trị **0** được xem như **thiếu dữ liệu**, vì trong thực tế không thể có:
- Glucose = 0
- BloodPressure = 0
- SkinThickness = 0
- Insulin = 0
- BMI = 0


```python
#Kiểm tra dữ liệu thiếu
print("Kiểm tra giá trị 0 (coi là thiếu):")
cols_with_zeros = ['Glucose','BloodPressure','SkinThickness','Insulin','BMI']
print((df[cols_with_zeros] == 0).sum())
```

3. Xử lý dữ liệu thiếu và EDA

- Thay thế các giá trị 0 bằng NaN để thuận tiện cho việc xử lý.

```python
df[cols_with_zeros] = df[cols_with_zeros].replace(0, np.nan)
print("Sau khi thay 0 bằng NaN:")
print(df.isna().sum())
```

4. Phân tích mô tả (EDA)

- Thực hiện phân tích trực quan để hiểu rõ đặc điểm dữ liệu.

- 4.1 Phân phối các biến quan trọng (Histogram)


```python
plt.figure(figsize=(12,8))
df[cols_with_zeros].hist(bins=20, figsize=(12,8), color='skyblue')
plt.suptitle('Histograms of Key Numerical Features', fontsize=16)
plt.show()
```

4.2 Boxplot – So sánh phân phối theo Outcome

- Dùng Boxplot để so sánh phân phối các biến giữa nhóm **Healthy (0)** và **Diabetic (1)**.


```python
# Boxplot để xem phân bố theo Outcome
for col in cols_with_zeros:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=df, x='Outcome', y=col, hue='Outcome', palette='coolwarm', legend=False)
    plt.title(f'{col} vs Outcome', fontsize=12)
    plt.xlabel('Outcome')
    plt.ylabel(col)
    plt.show()
```

5. Ma trận tương quan

- Kiểm tra mối tương quan giữa các biến và Outcome để xem biến nào quan trọng nhất.


```python
# Ma trận tương quan
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
